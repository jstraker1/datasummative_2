## 2) New Code

```python
def make_scatter(df,predictor,dependent_var):
    g = sns.lmplot(x=predictor,y=dependent_var,data=df,scatter_kws={'color': '#0000DD','alpha':0.025})
    g.ax.set_title(f'{predictor} vs {dependent_var}')
    g.ax.set_xlabel(predictor)
    g.ax.set_ylabel(dependent_var)
    return g

#def make_linear_regression(df,predictor,dependent_var):

#def make_logistic_regression(df,predictor,dependent_var):
```

```python
predictor = 'age'
dependent_var = 'bmi'
plt.show(make_scatter(df,predictor,dependent_var))
```
